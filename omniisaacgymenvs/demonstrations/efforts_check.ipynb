```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
from omniisaacgymenvs.demonstrations.demo_parser import load_dataset
```

```python
df = load_dataset()
```

```python
raw_torques = df["joint_effort"]
n = len(raw_torques)
torques = np.zeros((7, n))
for i in range(n):
    torques[:, i] = np.array(raw_torques[i][0])
```

```python
for i in range(7):
    plt.subplot(2,4, i+1)
    plt.title(f'action {i}')
    plt.plot(torques[i, :])
plt.show()
```
